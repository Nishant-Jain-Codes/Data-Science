```python
def printer(x):
    i = 0
    y=1
    while i < x:
        no = 3*y + 2
        if no % 4 != 0:
            print(no,end=" ")
            i += 1
        y += 1
x = int(input("enter x "))
printer(x)
```

```python
def sumN(n):
    i = 1
    sum = 0
    while i <= n:
        sum += i
        i += 1
    return sum
def prodN(n):
    i = 1
    prod = 1
    while i <= n:
        prod *= i
        i += 1
    return prod
n = int(input("enter n "))
choice = int(input("enter choice 1(sum) or 2(prod) "))
if choice == 1:
    print(sumN(n))
elif choice == 2:
    print(prodN(n))
else:
    print("invalid choice")
```

```python
def intRoot(x):
    ans = 0
    while(ans*ans <= x):
        ans += 1
    return ans-1
x = int(input("enter x "))
print(intRoot(x))
```

```python
def decimalToBinary(n):
    ans = 0
    i = 0
    while n > 0:
        ans += (n%2)*(10**i)
        n = n//2
        i += 1
    return ans
n = int(input("enter n "))
print(decimalToBinary(n))
```

```python
def binaryToDecimal(n):
    ans = 0
    i = 0
    while n > 0:
        ans += (n%10)*(2**i)
        n = n//10
        i += 1
    return ans
n = int(input("enter n "))
print(binaryToDecimal(n))
```

```python
def evenFibonacciNumbers(n):
    i = 1
    j = 1
    sum = 0
    while i <= n:
        if i % 2 == 0:
            sum += i
        k = i + j
        i = j
        j = k
    return sum
n = int(input("enter n "))
print(evenFibonacciNumbers(n))
```

```python
def trailingZeroesInNfactorial(n):
    i = 5
    count = 0
    while i <= n:
        count += n//i
        i *= 5
    return count
n = int(input("enter n "))
print(trailingZeroesInNfactorial(n))
```
