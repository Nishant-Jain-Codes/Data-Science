```python
def farenheitToCelcius(start,end,step):
    for i in range(start,end+1,step):
        print(i, int((i-32)*5/9))
start = int(input("enter start value: "))
end = int(input("enter end value: "))
step = int(input("enter step value: "))
farenheitToCelcius(start,end,step)
```

```python
def isFibonacciMember(n):
    a = 0
    b = 1
    while a < n:
        c = a + b
        a = b
        b = c
    if a == n:
        return True
    else:
        return False
n = int(input("enter a number: "))
isFibonacciMember(n)
```
