```python
def patternPrinter1(n):
    for i in range(0,n):
        for j in range(0,n):
            print(n,end='')
        print()
n = int(input("enter n "))
patternPrinter1(n)
```

```python
def patternPrinter2(n):
    for i in range(0,n):
        for j in range(0,i+1):
            print('*',end='')
        print()
n = int(input("enter n "))
patternPrinter2(n)
```

```python
def patternPrinter3(n):
    for i in range(1,n+1):
        for j in range(0,i):
            print(i,end='')
        print()
n = int(input("enter n "))
patternPrinter3(n)
```

```python
def patternPrinter4(n):
    for i in range(1,n+1):
        for j in range(i,0,-1):
            print(j,end='')
        print()
n = int(input("enter n "))
patternPrinter4(n)
```



The ord() function in Python is a built-in function that returns the Unicode code point of a given character. It takes a single character as an argument and returns an integer representing the Unicode code / ASCII point value of that character.

```python
def patternPrinter5(n):
    for i in range(0,n):
        for j in range(0,i+1):
            print(chr(ord('A') + i),end='')
        print()
n = int(input("enter n "))
patternPrinter5(n)
```

```python
def patternPrinter6(n):
    for i in range(0,n):
        for j in range(0,i+1):
            print(chr(ord('A') + j+i),end='')
        print()
n = int(input("enter n "))
patternPrinter6(n)
```

```python
def patternPrinter7(n):
    for i in range(n+1,0,-1):
        for j in range (i-1 , n , +1):
            print(chr(ord('A') + j),end='')
        print()
n = int(input("enter n "))
patternPrinter7(n)
```

```python
def patternPrinter8(n):
    for i in range(0,n+1):
        for j in range (0,n-i):
            print(' ',end='')
        for j in range (0,2*i + 1):
            print('*',end='')
        print()
n = int(input("enter n "))
patternPrinter8(n)
```

```python
def patternPrinter9(n):
    for i in range(1,n+1):
        for j in range (0,n-i):
            print(' ',end='')
        for j in range(i,2*i):
            print(j,end='')
        for j in range(2*i-2,i-1,-1):
            print(j,end='')
        print()
n = int(input("enter n "))
patternPrinter9(n)
```

```python
def patternPrinter10(n):
    for i in range(0,n+1):
        for j in range (0,n-i):
            print(' ',end='')
        for j in range (0,2*i + 1):
            print('*',end='')
        print()
    for i in range(n,-1,-1):
        for j in range (0,n-i):
            print(' ',end='')
        for j in range (0,2*i + 1):
            print('*',end='')
        print()
n = int(input("enter n "))
patternPrinter10(n)
```
