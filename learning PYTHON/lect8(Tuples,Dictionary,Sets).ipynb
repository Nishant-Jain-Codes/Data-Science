Tuple Methods
-------------

| Method | Description | Example |
| --- | --- | --- |
| `len()` | Returns the number of elements in the tuple | `len(my_tuple)` |
| `index(value)` | Returns the index of the first occurrence of `value` | `my_tuple.index(value)` |
| `count(value)` | Returns the number of occurrences of `value` | `my_tuple.count(value)` |

Dictionary Methods
------------------

| Method | Description | Example |
| --- | --- | --- |
| `len()` | Returns the number of key-value pairs in the dictionary | `len(my_dict)` |
| `keys()` | Returns a view object of all keys in the dictionary | `my_dict.keys()` |
| `values()` | Returns a view object of all values in the dictionary | `my_dict.values()` |
| `items()` | Returns a view object of all key-value pairs | `my_dict.items()` |
| `get(key, default=None)` | Returns the value for `key`, or `default` if key not found | `my_dict.get(key)` |
| `pop(key, default=None)` | Removes and returns the value for `key`, or `default` if key not found | `my_dict.pop(key)` |
| `popitem()` | Removes and returns a random key-value pair from the dictionary | `my_dict.popitem()` |
| `update(other_dict)` | Updates the dictionary with key-value pairs from `other_dict` | `my_dict.update(other_dict)` |
| `clear()` | Removes all key-value pairs from the dictionary | `my_dict.clear()` |

Set Methods
-----------

| Method | Description | Example |
| --- | --- | --- |
| `len()` | Returns the number of elements in the set | `len(my_set)` |
| `add(element)` | Adds an `element` to the set | `my_set.add(element)` |
| `remove(element)` | Removes an `element` from the set | `my_set.remove(element)` |
| `discard(element)` | Removes an `element` from the set if it is present | `my_set.discard(element)` |
| `pop()` | Removes and returns an arbitrary element from the set | `my_set.pop()` |
| `clear()` | Removes all elements from the set | `my_set.clear()` |
| `union(other_set)` | Returns a new set with elements from both sets | `my_set.union(other_set)` |
| `intersection(other_set)` | Returns a new set with elements common to both sets | `my_set.intersection(other_set)` |
| `difference(other_set)` | Returns a new set with elements in the first set but not in the second set | `my_set.difference(other_set)` |
| `symmetric_difference(other_set)` | Returns a new set with elements present in either set, but not in both | `my_set.symmetric_difference(other_set)` |

```python
def highestFrequencyNumber(arr):
    d = {}
    for i in arr:
        d[i] = d.get(i,0) + 1
    max = 0
    for i in arr:
        if d[i] > max:
            max = d[i]
            ans = i,d[i]
    return ans
arr = [int(x) for x in input("enter a list of elements ").split()]
print("highest frequency number :->  ",highestFrequencyNumber(arr))
```

```python
def maximumDistance(arr):
    d = {}
    for i in range(len(arr)):
        d[arr[i]] = i
    max = 0
    for i in range(len(arr)):
        if d[arr[i]] - i > max:
            max = d[arr[i]] - i
    return max
arr = [int(x) for x in input("enter a list of elements ").split()]
print("maximum distance :->  ",maximumDistance(arr))
```

```python
def differentNames(names):
    d = {}
    for i in names:
        d[i] = d.get(i,0) + 1
    answer = []
    for name,freq in d.items():
        if freq > 1:
            answer.append((name,freq))
    if len(answer) == 0:
        return -1
    else:
        return answer
names = [x for x in input("enter a list of names ").split()]
print("different names :->  ",differentNames(names))
```

```python
def evenCount(arr):
    d = {}
    for x in arr:
        d[x] = d.get(x,0) + 1
    for x in arr:
        if d[x] % 2 == 0:
            return x
    return -1
arr = [int(x) for x in input("enter a list of elements ").split()]
print("even count :->  ",evenCount(arr))
```

```python
def uniqueTriangles(triangles):
    d = {}
    for i in triangles:
        i.sort()
        d[tuple(i)] = d.get(tuple(i),0) + 1
    return len(d) , d.keys()
triangles = [[int(x) for x in input("enter a list of elements ").split()] for i in range(int(input("enter the number of triangles "))) ]
print("unique treiangles :->  ",uniqueTriangles(triangles))
```

```python
def pairsWithDifferenceK(arr, k):
    d = {}
    for i in arr:
        d[i] = d.get(i,0) + 1
    count = 0
    for i in arr:
        if k == 0:
            if d[i] > 1:
                count += 1
        else:
            if i + k in d:
                count += 1
    return count
k = int(input("enter the value of k "))
arr = [int(x) for x in input("enter a list of elements ").split()]
print("pairs with difference k :->  ",pairsWithDifferenceK(arr,k))
```
