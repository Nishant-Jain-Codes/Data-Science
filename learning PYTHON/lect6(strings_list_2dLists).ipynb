# strings methods
| Function | Description | Example |
| --- | --- | --- |
| `len()` | Returns the length of a string. | `len("Hello")` returns `5` |
| `lower()` | Converts a string to lowercase. | `"Hello".lower()` returns `"hello"` |
| `upper()` | Converts a string to uppercase. | `"Hello".upper()` returns `"HELLO"` |
| `capitalize()` | Capitalizes the first character of a string. | `"hello".capitalize()` returns `"Hello"` |
| `title()` | Converts the first character of each word to uppercase. | `"hello world".title()` returns `"Hello World"` |
| `count()` | Counts the occurrences of a substring in a string. | `"Hello".count("l")` returns `2` |
| `find()` | Searches for a substring in a string and returns its index. | `"Hello".find("lo")` returns `3` |
| `replace()` | Replaces occurrences of a substring with another substring. | `"Hello".replace("l", "X")` returns `"HeXXo"` |
| `split()` | Splits a string into a list of substrings based on a delimiter. | `"Hello,World".split(",")` returns `["Hello", "World"]` |
| `join()` | Joins a list of strings into a single string with a separator. | `" ".join(["Hello", "World"])` returns `"Hello World"` |
| `startswith()` | Checks if a string starts with a specified substring. | `"Hello".startswith("He")` returns `True` |
| `endswith()` | Checks if a string ends with a specified substring. | `"Hello".endswith("lo")` returns `True` |
| `strip()` | Removes leading and trailing whitespace from a string. | `" Hello ".strip()` returns `"Hello"` |
| `isdigit()` | Checks if a string consists of only digits. | `"123".isdigit()` returns `True` |
| `isalpha()` | Checks if a string consists of only alphabetic characters. | `"Hello".isalpha()` returns `True` |
| `islower()` | Checks if a string is in lowercase. | `"hello".islower()` returns `True` |
| `isupper()` | Checks if a string is in uppercase. | `"HELLO".isupper()` returns `True` |
| `isnumeric()` | Checks if a string is numeric. | `"123".isnumeric()` returns `True` |

List Operations
---------------

| Operation | Description | Example |
| --- | --- | --- |
| `len()` | Returns the length of a list. | `len([1, 2, 3])` returns `3` |
| `append()` | Adds an element to the end of a list. | `[1, 2, 3].append(4)` modifies the list to `[1, 2, 3, 4]` |
| `insert()` | Inserts an element at a specified index in a list. | `[1, 2, 3].insert(1, 4)` modifies the list to `[1, 4, 2, 3]` |
| `remove()` | Removes the first occurrence of an element from a list. | `[1, 2, 3, 4].remove(3)` modifies the list to `[1, 2, 4]` |
| `pop()` | Removes and returns the element at a specified index in a list. | `[1, 2, 3].pop(1)` modifies the list to `[1, 3]` and returns `2` |
| `index()` | Returns the index of the first occurrence of an element in a list. | `[1, 2, 3].index(2)` returns `1` |
| `count()` | Returns the number of occurrences of an element in a list. | `[1, 2, 2, 3].count(2)` returns `2` |
| `sort()` | Sorts the elements in a list in ascending order. | `[3, 1, 2].sort()` modifies the list to `[1, 2, 3]` |
| `reverse()` | Reverses the order of elements in a list. | `[1, 2, 3].reverse()` modifies the list to `[3, 2, 1]` |
| `copy()` | Returns a shallow copy of a list. | `new_list = old_list.copy()` |
| `clear()` | Removes all elements from a list. | `my_list.clear()` modifies the list to `[]` |

2D List Operations
------------------

| Operation | Description | Example |
| --- | --- | --- |
| Accessing Elements | Accessing elements in a 2D list using row and column indices. | `matrix = [[1, 2], [3, 4]]`<br>`matrix[0][1]` returns `2` |
| Iterating | Iterating through elements in a 2D list using nested loops. | `matrix = [[1, 2], [3, 4]]`<br>`for row in matrix:`<br>    `for element in row:`<br>        `print(element)` |
| Modifying Elements | Modifying elements in a 2D list using row and column indices. | `matrix = [[1, 2], [3, 4]]`<br>`matrix[1][0] = 5` modifies the list to `[[1, 2], [5, 4]]` |
| Adding Rows | Adding a new row to a 2D list. | `matrix = [[1, 2], [3, 4]]`<br>`matrix.append([5, 6])` modifies the list to `[[1, 2], [3, 4], [5, 6]]` |
| Removing Rows | Removing a row from a 2D list. | `matrix = [[1, 2], [3, 4], [5, 6]]`<br>`matrix.pop(1)` modifies the list to `[[1, 2], [5, 6]]` |
| Adding Columns | Adding a new column to a 2D list. | `matrix = [[1, 2], [3, 4]]`<br>`for row in matrix:`<br>    `row.append(0)` modifies the list to `[[1, 2, 0], [3, 4, 0]]` |
| Removing Columns | Removing a column from a 2D list. | `matrix = [[1, 2, 0], [3, 4, 0]]`<br>`for row in matrix:`<br>    `row.pop(1)` modifies the list to `[[1, 0], [3, 0]]` |
| Copying | Creating a copy of a 2D list. | `new_matrix = [row[:] for row in matrix]` creates a copy of the 2D list `matrix` |
| Nested List Comprehension | Creating a new 2D list using a nested list comprehension. | `matrix = [[i+j for i in range(3)] for j in range(3)]` creates a 3x3 matrix with values `[[0, 1, 2], [1, 2, 3], [2, 3, 4]]` |

```python
def isPalindrome(str):
    for i in range(len(str)//2):
        if str[i] != str[-i-1]:
            return False
    return True
str = input("enter a string: ")
print(isPalindrome(str))
```

```python
def printAllSubstring(str):
    for i in range(len(str)):
        for j in range(i+1,len(str)+1):
            print(str[i:j])
str = input("enter a string: ")
printAllSubstring(str)
```

```python
def swapAlternate(arr):
    for i in range(0,len(arr)-1,2):
        arr[i],arr[i+1] = arr[i+1],arr[i]
    return arr
arr = [int(x) for x in input("enter a list of numbers: ").split()]
print(swapAlternate(arr))
```

```python
def arrayEquilibriumIndex(arr):
    first = 0
    last = sum(arr)  
    for i in range(len(arr)):
        last -= arr[i]  
        if first == last:
            return i
        first += arr[i]  
    return -1
arr = [int(x) for x in input("enter a list of numbers: ").split()]
print(arrayEquilibriumIndex(arr))
```

```python
def rowWiseSum(mat):
    for i in range(len(mat)):
        sum = 0
        for j in range(len(mat[i])):
            sum += mat[i][j]
        print(sum,end=" ")
mat = [[int(x) for x in input("enter a list of numbers: ").split()] for i in range(int(input("enter no. of rows: ")))]
rowWiseSum(mat)
```

```python
def wavePrint(mat):
    for i in range(len(mat[0])):
        if i%2 == 0:
            for j in range(len(mat)):
                print(mat[j][i],end=" ")
        else:
            for j in range(len(mat)-1,-1,-1):
                print(mat[j][i],end=" ")
mat = [[int(x) for x in input("enter a list of numbers: ").split()] for i in range(int(input("enter no. of rows: ")))]
wavePrint(mat)
```

```python
def removeConsecutiveDuplicates(str):
    ans = str[0]
    for i in range(1,len(str)):
        if str[i] != str[i-1]:
            ans += str[i]
    return ans
str = input("enter a string: ")
print(removeConsecutiveDuplicates(str))
```

```python
def reverseEachWord(str):
    ans = ""
    for i in str.split():
        ans += i[::-1] + " "
    return ans
str = input("enter a string: ")
print(reverseEachWord(str))
```

```python
def highestOccuringCharacter(str):
    freq = {}
    for i in str:
        if i in freq:
            freq[i] += 1
        else:
            freq[i] = 1
    max = 0
    for i in freq:
        if freq[i] > max:
            max = freq[i]
            ans = i
    return ans
str = input("enter a string: ")
print(f"hightest occuring character is: '{highestOccuringCharacter(str)}'")
```

```python
def arrangeNumbersInArray(n):
    arr = []
    for i in range(0,(n+1)//2 ):
        arr.append(i*2 + 1)
    for i in range(n//2,0,-1):
        arr.append(i*2)
    return arr
n = int(input("enter a number: "))
print(arrangeNumbersInArray(n))
```

```python
def leadersInArray(arr):
    ans = []
    for i in range(len(arr)):
        flag = True
        for j in range(i+1,len(arr)):
            if arr[i] < arr[j]:
                flag = False
                break
        if flag:
            ans.append(arr[i])
    return ans
arr = [int(x) for x in input("enter a list of numbers: ").split()]
print(leadersInArray(arr))
```

```python
def maximiseTheSum(arr1,arr2):
    
    i = 0
    j = 0
    sum1 = 0
    sum2 = 0
    max_sum = 0
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            sum1 += arr1[i]
            i += 1
        elif arr1[i] > arr2[j]:
            sum2 += arr2[j]
            j += 1
        else:
            max_sum += max(sum1, sum2) + arr1[i]
            sum1 = 0
            sum2 = 0
            i += 1
            j += 1
    while i < len(arr1):
        sum1 += arr1[i]
        i += 1
    while j < len(arr2):
        sum2 += arr2[j]
        j += 1
    max_sum += max(sum1, sum2) 
    return max_sum
arr1 = [int(x) for x in input("enter a list of numbers: ").split()]
arr2 = [int(x) for x in input("enter a list of numbers: ").split()]
print(maximiseTheSum(arr1,arr2))
```

```python
def largestRowOrColumn(mat):
    max = 0
    for i in range(len(mat)):
        sum = 0
        for j in range(len(mat[i])):
            sum += mat[i][j]
        if sum > max:
            max = sum
            ans = "row",i,max
    for i in range(len(mat[0])):
        sum = 0
        for j in range(len(mat)):
            sum += mat[j][i]
        if sum > max:
            max = sum
            ans = "column",i,max
    return ans
mat = [[int(x) for x in input("enter a list of numbers: ").split()] for i in range(int(input("enter no. of rows: ")))]
print(largestRowOrColumn(mat))
```

```python
def spiralPrint(mat):
    top = 0
    bottom = len(mat)-1
    left = 0
    right = len(mat[0])-1
    while top <= bottom and left <= right:
        for i in range(left,right+1):
            print(mat[top][i],end=" ")
        top += 1
        for i in range(top,bottom+1):
            print(mat[i][right],end=" ")
        right -= 1
        if top <= bottom:
            for i in range(right,left-1,-1):
                print(mat[bottom][i],end=" ")
            bottom -= 1
        if left <= right:
            for i in range(bottom,top-1,-1):
                print(mat[i][left],end=" ")
            left += 1
mat = [[int(x) for x in input("enter a list of numbers: ").split()] for i in range(int(input("enter no. of rows: ")))]
spiralPrint(mat)
```
