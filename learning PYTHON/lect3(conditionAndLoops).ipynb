Write a program that performs the tasks of a simple calculator. The program should first take an integer as input and then based on that integer perform the task as given below.
1. If the input is 1, then 2 integers are taken from the user and their sum is printed.
2. If the input is 2, then 2 integers are taken from the user and their difference(1st number - 2nd number) is printed.
3. If the input is 3, then 2 integers are taken from the user and their product is printed.
4. If the input is 4, then 2 integers are taken from the user and the quotient obtained (on dividing 1st number by 2nd number) is printed.
5. If the input is 5, then 2 integers are taken from the user and their remainder(1st number mod 2nd number) is printed.
6. If the input is 6, then the program exits.
7. For any other input, then print "Invalid Operation".


```python
def sum(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def rem(a,b):
    return a%b
def div(a,b):
    return a/b

a = int(input("enter a"))
b = int(input("enter b"))
choice = int(input("enter choice"))

if(choice==1):
    print(sum(a,b))
elif(choice==2):
    print(sub(a,b))
elif(choice==3):
    print(mul(a,b))
elif(choice==4):
    print(rem(a,b))
elif(choice==5):
    print(div(a,b))
else:
    print("invalid choice")
```

```python
def reverse(no):
    rev = 0
    while(no>0):
        rem = no%10
        rev = rev*10 + rem
        no = no//10
    return rev  
no = int(input("enter no"))
print(f"reverse no is {reverse(no)}")
```



The [::-1] part is called slicing with a step value of -1. It specifies that you want to traverse the string from the last character to the first character, in reverse order. If you omit the first two values, it means start from the beginning of the string, and if you omit the last value, it means stop at the end of the string.

```python
def isPalindrome(str):
    rev = str[::-1]
    if(rev==str):
        return True
    else:
        return False
str = input("enter str")
print(isPalindrome(str))
```

```python
def sumOddEven(no):
    odd = 0
    even = 0
    while(no):
        curNo = no%10
        if(curNo%2==0):
            even+=curNo
        else:
            odd+=curNo
        no//=10
    print(f"sum of odd is {odd} and sum of even is {even}")
no = int(input("enter no"))
sumOddEven(no)
```

```python
def nthFib(n):
    if(n==1):
        return 0
    elif(n==2):
        return 1
    else:
        return nthFib(n-1)+nthFib(n-2)
n = int(input("enter n"))
print(f"{n}th fib is {nthFib(n)}")
```
